```python
import pandas as pd
with open('D:\GB\Сбор и разметка данных\killers_by_countries.csv', 'r') as f:
    df = pd.read_csv(f, header=None, skiprows=1)

# Задаю имена столбцам
df.columns = ['position', 'name', 'kills_qty', 'year']
```

```python
df18 = df[df['year'] == 2018]
```

```python
df13 = df[df['year'] == 2013]
df13 = df13.sort_values('kills_qty', ascending=False)
df13['position'] = df13['kills_qty'].rank(ascending=False, method='min').astype(int)
df13
```

К великому сожалению, одни и те же страны в обоих таблицах называются по-разному, чтобы корректно соединить таблицы по столбцам, пришлось вручную обработать эти несовпадения и некоторые названия заменить.

```python
replace13 = {'Американские Виргинские Острова': 'Виргинские о-ва (США)',
    'Британские Виргинские Острова': 'Виргинские о-ва (Брит.)',
    'Сент-Винсент': 'Сент-Винсент и Гренадины',
    'Южная Африка': 'ЮАР',
    'Центрально-Африканская Республика': 'ЦАР',
    'Сейшельские Острова': 'Сейшельские о-ва',
    'Коста Рика': 'Коста-Рика',
    'Папуа — Новая Гвинея': 'Папуа - Новая Гвинея',
    'Каймановы острова': 'Каймановы о-ва',
    'Соединенные Штаты Америки': 'США',
    'Тимор-Лешти': 'Восточный Тимор',
    'Маршалловы Острова': 'Маршалловы о-ва',
    'Буркина Фасо': 'Буркина-Фасо',
    'Македония': 'Северная Македония',
    'Объединенные Арабские Эмираты': 'ОАЭ'}
replace18 = {'Багамы': 'Багамские Острова',
    'Острова Теркс и Кайкос': 'Тёркс и Кайкос',
    'Американское Самоа': 'Самоа',
    'Киргизия': 'Кыргызстан'}

df13['name'] = df13['name'].replace(replace13)
df18['name'] = df18['name'].replace(replace18)
```

```python
kills_dif = df13.merge(df18, how = 'left', on = 'name')
kills_dif
```

Такая таблица показалась мне не очень удобной, к тому же метка по году нам теперь не нужна, поэтому далее я немного меняю порядок столбцов, их количество и названия.

```python
kills_dif = kills_dif.rename(columns={'position_x': 'position13', 'name': 'name', 'kills_qty_x': 'kills_qty13', 'year_x': 'year13', 'position_y': 'position18', 'kills_qty_y': 'kills_qty18', 'year_y': 'year18'})
kills_dif = kills_dif[['name','position13', 'kills_qty13', 'year13', 'position18', 'kills_qty18', 'year18']]
kills_dif = kills_dif.drop(columns = ['year18', 'year13'],axis = 1)

kills_dif
```

Далее я хочу посчитать динамику, как менялся рейтинг страны по количеству преднамеренных убийств на 100 тыс. населения и, собственно, число этих убийств.

```python
kills_dif['rate_dif'] = kills_dif['position18'] - kills_dif['position13']
kills_dif['qty_dif'] = kills_dif['kills_qty18'] - kills_dif['kills_qty13']
kills_dif
```

Сделала дф со странами, где рейтинг уменьшается, то есть становится безопаснее, и со странами, где, наоборот, рейтинг растет.

```python
impr_countries = kills_dif[(kills_dif['rate_dif'] < 0)]
impr_countries
```

```python
worse_countries = kills_dif[(kills_dif['rate_dif'] > 0)]
worse_countries
```
